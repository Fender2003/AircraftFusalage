```python
import os

from ultralytics import YOLO
import cv2

model = YOLO('/content/gdrive/My Drive/aircraft_fuselage_yolo/runs/detect/train2/weights/last.pt')

img_path = '/content/gdrive/My Drive/aircraft_fuselage_yolo/unlabel_aircraft_fuselage/1338.jpg'
image = cv2.imread(img_path)

results = model(image)[0]

threshold = 0.5

for result in results.boxes.data.tolist():
    x1, y1, x2, y2, score, class_id = result

    if score > threshold:
        cv2.rectangle(image, (int(x1), int(y1)), (int(x2), int(y2)), (0, 255, 0), 4)
        cv2.putText(image, results.names[int(class_id)].upper(), (int(x1), int(y1 - 10)),
                    cv2.FONT_HERSHEY_SIMPLEX, 1.3, (0, 255, 0), 3, cv2.LINE_AA)

cv2.imwrite('/content/gdrive/My Drive/aircraft_fuselage_yolo/Predicted_Images/1338.jpg', image)
# cv2.imshow('Output', image)
cv2.waitKey(0)
```

```python
# VIDEOS_DIR = os.path.join('.', 'videos')
import cv2
from ultralytics import YOLO
video_path = '/content/gdrive/My Drive/aircraft_fuselage_yolo/unlabel_aircraft_fuselage/invideo-ai-1080 The Unseen Dangers_ Aircraft Wing Cracks 2024-01-29.mp4'
video_path_out = '/content/gdrive/My Drive/aircraft_fuselage_yolo/Predicted_Images/vieo_out.mp4'

cap = cv2.VideoCapture(video_path)
cap.open(video_path)

ret, frame = cap.read()
H, W, _ = frame.shape
out = cv2.VideoWriter(video_path_out, cv2.VideoWriter_fourcc(*'MP4V'), int(cap.get(cv2.CAP_PROP_FPS)), (W, H))

model = YOLO('/content/gdrive/My Drive/aircraft_fuselage_yolo/runs/detect/train2/weights/last.pt')

threshold = 0.5

while ret:

    results = model(frame)[0]

    for result in results.boxes.data.tolist():
        x1, y1, x2, y2, score, class_id = result

        if score > threshold:
            cv2.rectangle(frame, (int(x1), int(y1)), (int(x2), int(y2)), (0, 255, 0), 4)
            cv2.putText(frame, results.names[int(class_id)].upper(), (int(x1), int(y1 - 10)),
                        cv2.FONT_HERSHEY_SIMPLEX, 1.3, (0, 255, 0), 3, cv2.LINE_AA)

    out.write(frame)
    ret, frame = cap.read()

cap.release()
out.release()
cv2.destroyAllWindows()
```
